```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#read data
names1880 = pd.read_csv('names/yob1880.txt', names=['names', 'sex', 'births'])
names1880
```

```python
#how many children was born in 1880
names1880.groupby('sex').births.sum()
```

```python
#concat all years
years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'names/yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)
    
    names = pd.concat(pieces, ignore_index=True)
    
names    
```

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
total_births.tail()
```

```python
total_births.plot(title='Total birth by sex and year')
```

```python
#add column prop: births/sum of births
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
names
```

```python
#check prop
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
#extract first 1000 names for each combination of year and sex
def get_top1000(group):
    return group.sort_index(by='births', ascending=False) [:1000]
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000
```

```python
#divide top 1000 on two parts: boys and girls
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']

#will form general table where will shows general count was birthed by years and names
total_births = top1000.pivot_table('births', index='year', columns='name', aggfunc=sum)
total_births
```

```python
#will build graphics
subset = total_births[['Ivan', 'Ali', 'Pat']]
subset.plot(subplots=True, figsize=(12, 10), grid=False, title='Number of births per year')
```

```python
#The diversity of growth measurement
table = top1000.pivot_table('prop', index='year', columns='sex', aggfunc=sum)
table.plot(title='Sum of top1000.prop by year and sex', yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2010, 10))
```

```python
df = boys[boys.year == 2010]
df
```

```python
prop_cumsum = df.sort_index(by='prop', ascending=False).prop.cumsum()
prop_cumsum[:10]
```

```python
#50% boys has been named one of the 117 names in 2010 
prop_cumsum.searchsorted(0.5)
```

```python
#50% boys has been named one of the 25 names in 1900
df = boys[boys.year == 1900]
in1900 = df.sort_index(by='prop', ascending=False).prop.cumsum()
in1900.searchsorted(0.5) + 1
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()
```

```python
#"The last letter" of revolution
#extract the last letter in the column name
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letters'
table = names.pivot_table('births', index=last_letters, columns=['sex', 'year'], aggfunc=sum)
subtable = table.reindex(columns=[1910,1960,2010],level='year')
subtable.head()
```

```python
#normalize
subtable.sum()
```

```python
letter_prop = subtable / subtable.sum().astype(float)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10,8))
letter_prop['M'].plot(kind='bar', rot=0, ax=axes[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axes[1], title='Female', legend=False)
```

```python
letter_prop = table / table.sum().astype(float)
dny_ts = letter_prop.ix[['d', 'n', 'y'], 'M'].T
dny_ts.head()
```

```python
#men's names have become women and vice versa
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births', index='year', columns='sex', aggfunc=sum)
table = table.div(table.sum(1), axis=0)
table.tail()
```

```python
table.plot(style={'M':'k-', 'F':'k--'})
```
